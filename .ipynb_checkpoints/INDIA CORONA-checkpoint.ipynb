```python
import pandas as pd
import numpy as np
CoronaSummary=pd.read_excel("D:\\jupyter\\india corona\\india corona.xls",index_col="State/Union Territory")
CoronaSummary.head()
```

```python
CoronaStateCase=CoronaSummary.drop(labels='S.No.',axis=1)#['S.No.']
CoronaStateCase.head()
```

```python
StateTotalCases=CoronaStateCase.iloc[:,0]
StateTotalCasesLimit=StateTotalCases[StateTotalCases>50]
#StateTotalCases.ix[(StateTotalCases>50)]
StateTotalCasesLimit.plot.bar(x='State/Union Territory',y='total cases',width=.5
                              ,title='Case Above 50')
```

```python
StateDR=CoronaStateCase.iloc[:,1:3]
StateDRLimit=StateDR[(StateDR['Deaths']>0)&(StateDR['Recoveries']>0)]
StateDRLimit
StateDRLimit.plot()
```

```python
CoronaTotal=CoronaStateCase.sum()
CoronaTotal.plot.bar()
```

```python
StateTDR=CoronaStateCase.iloc[:,0:3]
StateTDR.head()
```

```python
def f(row):
    if (row['Total cases'] -(row['Deaths']+row['Recoveries']))>100:
        val = 'critical'
    else:
        val = 'less critical'
    return val
```

```python
StateTDR['Status'] = StateTDR.apply(f, axis=1)
StateTDR.head(2)
```

```python
StateTDR['ActiveCases']=(StateTDR['Total cases'] -(StateTDR['Deaths']+StateTDR['Recoveries']))
StateTDR.head()
```

```python
StateTDRActive=StateTDR.iloc[:,3:5]
StateTDRActiveLimit=StateTDRActive[(StateTDRActive['ActiveCases']>500)]
StateTDRActiveLimit
StateTDRActiveLimit.plot()
```

```python
Status=StateTDR.iloc[:,3:5]
#a.sort_values(by='ActiveCases')
GroupStatus=Status.groupby("Status")
GroupStatus.sum()
```

```python
CoronaCaseSorted=CoronaStateCase.sort_values("Active cases",ascending=False)
CoronaCaseSorted.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import HTML
```

```python
def color_negative_red(value):
  if value > 100:
    color = 'red'
  elif (50<value < 100):
    color = 'black'
  elif value < 50:
    color = 'green'

  return 'color: %s' % color
```

```python
CoronaCaseSortedColor=CoronaCaseSorted.style.applymap(color_negative_red)#.background_gradient()
CoronaCaseSortedColor
```

```python
CoronaDelhi=CoronaCaseSorted.iloc[1]
CoronaDelhi
```
